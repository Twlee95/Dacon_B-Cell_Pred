## number_of_tested and number_of_responses

```python
## number_of_tested, number_of_responses 두가지 변수는 결측치 비율이 80%에 가까움
print(len(lg_data.loc[:,"number_of_tested"]))
print(lg_data.loc[:,"number_of_tested"].isnull().sum())
print(lg_data.loc[:,"number_of_tested"].isnull().sum()/len(lg_data.loc[:,"number_of_tested"]))

print(len(lg_data.loc[:,"number_of_responses"]))
print(lg_data.loc[:,"number_of_responses"].isnull().sum())
print(lg_data.loc[:,"number_of_responses"].isnull().sum()/len(lg_data.loc[:,"number_of_responses"]))
```

```python
# Ttest_indResult(statistic=-40.94217381143185, pvalue=0.0) 
import matplotlib.pyplot as plt

aa= lg_data.loc[lg_data.loc[:,"label"]==0,["number_of_responses","label"]].dropna()

bb= lg_data.loc[lg_data.loc[:,"label"]==1,["number_of_responses","label"]].dropna()

print(aa.loc[:,"number_of_responses"].mean())
print(bb.loc[:,"number_of_responses"].mean())

plt.scatter(aa.loc[:,"number_of_responses"],aa.loc[:,"label"])

import scipy.stats

scipy.stats.ttest_ind(aa.loc[:,"number_of_responses"], bb.loc[:,"number_of_responses"], equal_var=False) 
```

```python
list1 = lg_data['assay_method_technique'].unique().tolist()
assay_method_technique_label = lg_data['assay_method_technique'].map(lambda x: list1.index(x))
list2 = lg_data['assay_group'].unique().tolist()
assay_group_label = lg_data['assay_group'].map(lambda x: list2.index(x))
list3 = lg_data['disease_type'].unique().tolist()
disease_type_label = lg_data['disease_type'].map(lambda x: list3.index(x))
list4 = lg_data['disease_state'].unique().tolist()
disease_state_label = lg_data['disease_state'].map(lambda x: list4.index(x))
```
